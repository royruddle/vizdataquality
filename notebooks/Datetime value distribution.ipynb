# Date/time value distribution at multiple levels of detail

This example shows how multiple plots are needed to show the distribution of dates/times at different levels of detail.

## Includes: viz_data_quality and other libraries

```python
import pandas as pd
import numpy as np
import datetime
import matplotlib as mpl
import matplotlib.pyplot as plt

from vizdataquality import calculate as vduc, datasets as vdqd, plot as vdqp
```

## Get a dataframe's profile

```python
num_rows, df = vdqd.get_dataset('datetime 1')
df_output = vdqd.calc(df)
print('Number of rows =', num_rows)
```

## Visualize the number of times each value occurs

### The default plot selects a level of detail automatically

```python
vdqp.datetime_counts(df['Date'])
```

### Different patterns are shown if each level of detail is plotted separately

Year: The data spans September 2022 to August 2023, so there are more values for 2023 than 2022.

Month: A slightly uneven distribution, caused by the number of days in each month.

Day of the week: The data are for 365 days. There is one more Thursday than any other day of the week.

Hour: A sinusoidal distribution for part of the day.

Minute: The same number of values for each minute.

Second: The values are only for zero and 30 seconds past the minute.

```python
fig_kw = {'size_inches': (9, 5), 'constrained_layout': True}
components = ['year', 'month', 'dayofweek', 'hour', 'minute', 'second']
vdqp.plotgrid('datetime distribution', df['Date'], num_rows=2, num_cols=3, vert=False, xlabels_rotate=[0, 90, 0, 0, 0, 0], fig_kw=fig_kw, components=components)
```

### This example shows the overall data (left) and number of records in each year (right)

```python
fig_kw = {'size_inches': (9, 3), 'constrained_layout': True}
components = [None, 'year']
vdqp.plotgrid('datetime distribution', df['Date'], num_cols=2, vert=False, xlabels_rotate=[45, 0], fig_kw=fig_kw, components=components)
```
