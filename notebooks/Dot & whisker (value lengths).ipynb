# Dot & whisker plots

This example shows how dot & whisker plots can be used to visualize the number of characters in the values of variables (this is known as value lengths).

## Includes: viz_data_quality and other libraries

```python
import pandas as pd
import numpy as np
import datetime
import matplotlib as mpl
import matplotlib.pyplot as plt

from vizdataquality import calculate as vdqc, datasets as vdqd, plot as vdqp
```

## Get a dataframe's profile

```python
num_rows, df = vdqd.get_dataset('simple')
df_output = vdqc.calc(df)
print('Number of rows =', num_rows)
print()
column = 'Value lengths'
print(column, '(comma-separated min/max value lengths)')
print(df_output[column])
```

## Visualize the value lengths of all the variables

```python
vdqp.dot_whisker(df_output[column])
```

## Customise the visualization

```python
# A dictionary of keyword arguments that are passed to the Matplotlib Figure object
fig_kw = {'size_inches': (6, 2), 'dpi': 300, 'constrained_layout': True}
# A dictionary of keyword arguments that are passed to the Matplotlib Axes object
ax_kw = {'title': 'Value lengths of each field', 'xlabel': 'Field', 'ylabel': 'Number of characters', 'ylim': (0, 20)}
##### A dictionary of keyword arguments that are passed to the Matplotlib bar or barh object
####kwargs = {'color': 'orange'}
# Rotate the X axis labels, save the plot to a file and apply the keyword arguments
vdqp.dot_whisker(df_output[column], xlabels_rotate=90, filename='value lengths.jpg', fig_kw=fig_kw, ax_kw=ax_kw)
```

## Horizontal rather than vertical plot

```python
vdqp.dot_whisker(df_output[column], vert=False)
```

## The variables can be divided into multiple plots (useful if there are many variables)

### Multiple rows of plots

```python
plt_kw = {'ylim': 0}
vdqp.multiplot('dot-or-whisker', df_output[column], number_of_variables_per_row=3, plt_kw=plt_kw)
```

### Multiple columns of plots

```python
fig_kw = {'constrained_layout': True}
plt_kw = {'xlim': 0}
vdqp.multiplot('dot-or-whisker', df_output[column], number_of_variables_per_row=3, vert=False, fig_kw=fig_kw, plt_kw=plt_kw)
```
