# Textual data

This example shows how textual data can be visualized within a plot.

## Includes: viz_data_quality and other libraries

```python
import pandas as pd
import numpy as np
import datetime
import matplotlib as mpl
import matplotlib.pyplot as plt

from vizdataquality import calculate as vdqc, datasets as vdqd, plot as vdqp
```

## Get a dataframe's profile

```python
num_rows, df = vdqd.get_dataset('simple')
df_output = vdqc.calc(df)
print('Number of rows =', num_rows)
column = 'Data type'
print()
print(column)
print(df_output[column])
```

## Visualize text

```python
vdqp.text(df_output[column])
```

## Customise the visualization

```python
# A dictionary of keyword arguments that are passed to the Matplotlib Figure object
fig_kw = {'size_inches': (1.5, 1.1), 'dpi': 300}
vdqp.text(df_output[column], filename='missing values.jpg', overwrite=False, fig_kw=fig_kw)
```
