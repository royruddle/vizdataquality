# Box plots

This example shows how box plots can be used to visualize the distribution (including outliers) of numerical variables.

## Includes: viz_data_quality and other libraries

```python
import pandas as pd
import numpy as np
import datetime
import matplotlib as mpl
import matplotlib.pyplot as plt

from vizdataquality import calculate as vdqc, datasets as vdqd, plot as vdqp
```

## Get a dataframe's profile

```python
num_rows, df = vdqd.get_dataset('numeric 2')
```

## Visualize the distribution of all the variables

```python
vdqp.boxplot_raw(df)
```

## Horizontal rather than vertical bars, which is also customised

```python
fig_kw = {'size_inches': (6, 3), 'dpi': 300, 'constrained_layout': True}
ax_kw = {'xlabel': 'Number'}
vdqp.boxplot_raw(df, vert=False, xlabels_rotate=45, filename='distribution boxplot.jpg', fig_kw=fig_kw, ax_kw=ax_kw)
```

## The variables can be divided into multiple plots (useful if there are many variables)

### Multiple rows of plots

NB: To create multiple columns of plots instead, set vert=False

```python
fig_kw = {'constrained_layout': True}
vdqp.multiplot('boxraw', df, number_of_variables_per_row=2, fig_kw=fig_kw)
```

### A grid of plots can be visualized, to show a boxplot for each variable

```python
fig_kw = {'size_inches': (6, 4), 'constrained_layout': True}
ax_kw = {'ylabel': None}
vdqp.plotgrid('boxraw', df, num_cols=2, vert=False, fig_kw=fig_kw, ax_kw=ax_kw)
```
