# Histograms (the distribution of numerical values)

This example shows how histograms can be used to visualize the distribution of values in each numerical variable.

## Includes: viz_data_quality and other libraries

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

from vizdataquality import datasets as vdqd, plot as vdqp
```

## Get a dataframe's profile

```python
num_rows, df = vdqd.get_dataset('numeric 2')
print('The variables are:', df.columns)
```

## First, here is a Matplotlib histogram

Notice how the right hand side of the plot is empty. The last bars are too short to be seen.

```python
plt.hist(df['Exponential'])
plt.show()
```

## vizdataquality histogram

This uses perceptual discontuity to ensure every bar is visible.

```python
vdqp.histogram(df['Exponential'], perceptual_threshold=0.01, datalabels=True)
```

## Horizontal rather than vertical plot, which is also customised

```python
# A dictionary of keyword arguments that are passed to the Matplotlib Figure object
fig_kw = {'size_inches': (4, 3), 'dpi': 300, 'constrained_layout': True}
# A dictionary of keyword arguments that are passed to the Matplotlib Axes object
ax_kw = {'title': 'Histogram', 'xlabel': 'Number of values'}
vdqp.histogram(df['Exponential'], perceptual_threshold=0.01, vert=False, datalabels=True,
               filename='distribution histogram.jpg', fig_kw=fig_kw, ax_kw=ax_kw)
```

## A grid of plots can be visualized, to show a histogram for each variable

```python
fig_kw = {'size_inches': (6, 4), 'constrained_layout': True}
ax_kw = {'title': df.columns.tolist()}
vdqp.plotgrid('histogram', df, num_cols=2, fig_kw=fig_kw, ax_kw=ax_kw)
```
