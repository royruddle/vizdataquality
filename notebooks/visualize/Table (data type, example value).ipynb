# Textual data

This example shows how textual data can be plotted in a table.

## Includes: viz_data_quality and other libraries

```python
import pandas as pd
import numpy as np
import datetime
import matplotlib as mpl
import matplotlib.pyplot as plt

from vizdataquality import calculate as vdqc, datasets as vdqd, plot as vdqp
```

## Get a dataframe's profile

```python
num_rows, df = vdqd.get_dataset('simple')
df_output = vdqc.calc(df)
print('Number of rows =', num_rows)
columns = ['Data type', 'Example value']
print()
print(df_output[columns])
```

## Plot the table

```python
# The variable names are stored in the index of df_output.
# The whitespace above the table is a 'feature' of matplotlib.axes.Axes.table
vdqp.table(df_output[columns], include_index=True)
```

## Customise the visualization

```python
# A dictionary of keyword arguments that are passed to the Matplotlib Figure object
fig_kw = {'size_inches': (4, 2), 'dpi': 300}
# Turn off automatic resizing of the columns, but put the table in a central location.
vdqp.table(df_output[columns], include_index=True, auto_column_width=False, filename='table.jpg', overwrite=False, fig_kw=fig_kw, loc='center')
```
