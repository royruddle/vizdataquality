# Violin plots (the distribution of numerical values)

This example shows how violin plots can be used to visualize the distribution of values in each numerical variable.

## Includes: viz_data_quality and other libraries

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

from vizdataquality import datasets as vdqd, plot as vdqp
```

## Get a dataframe's profile

```python
num_rows, df = vdqd.get_dataset('numeric 2')
```

## Visualize the distributions

```python
vdqp.violinplot(df)
```

## Customise the visualization

```python
# A dictionary of keyword arguments that are passed to the Matplotlib Figure object
fig_kw = {'size_inches': (2, 4), 'dpi': 300, 'constrained_layout': True}
# A dictionary of keyword arguments that are passed to the Matplotlib Axes object
ax_kw = {'title': 'Distribution', 'xlabel': 'Dataset', 'ylabel': 'Count'}

# Rotate the X axis labels, save the plot to a file and apply the keyword arguments.
# widths, showmedians and showextrema are some of the Matplotlib violinplot object keywords.
vdqp.violinplot(df, xlabels_rotate=90, filename='distribution violin.jpg', fig_kw=fig_kw, ax_kw=ax_kw,
                widths=0.75, showmedians=True, showextrema=False)
```

## Horizontal rather than vertical plot

```python
vdqp.violinplot(df, vert=False)
```

## The variables can be divided into multiple plots (useful if there are many variables)

### Multiple rows of plots

```python
vdqp.multiplot('violin', df, number_of_variables_per_row=3)
```

### Multiple columns of plots

```python
# A dictionary of keyword arguments that are passed to the Matplotlib Figure object
fig_kw = {'constrained_layout': True}
vdqp.multiplot('violin', df, number_of_variables_per_row=3, fig_kw=fig_kw, vert=False)
```
